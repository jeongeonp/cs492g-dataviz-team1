```python
physical_agg['z_calories'] = (physical_agg['p3012caloriesToday'] - physical_agg['calories_mean']) / physical_agg['calories_std']
physical_agg['z_pedometer'] = (physical_agg['p3012pedometerToday'] - physical_agg['pedometer_mean']) / physical_agg['pedometer_std']
physical_agg['z_physical'] = (physical_agg['z_calories'] + physical_agg['z_pedometer']) / 2

physical_agg
```
