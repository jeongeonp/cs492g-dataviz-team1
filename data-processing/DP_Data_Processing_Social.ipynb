```python
# for social agg
social_3005 = pd.read_csv('/content/drive/My Drive/CS492 DP/Data processing/[final] social/social-p3005.csv')
social_3010 = pd.read_csv('/content/drive/My Drive/CS492 DP/Data processing/[final] social/social-p3010.csv')
social_3011 = pd.read_csv('/content/drive/My Drive/CS492 DP/Data processing/[final] social/social-p3011.csv')
social_3012 = pd.read_csv('/content/drive/My Drive/CS492 DP/Data processing/[final] social/social-p3012.csv')
social_3030 = pd.read_csv('/content/drive/My Drive/CS492 DP/Data processing/[final] social/social-p3030.csv')

social_3005.set_index('datetime', drop=True, inplace=True)
social_3010.set_index('datetime', drop=True, inplace=True)
social_3011.set_index('datetime', drop=True, inplace=True)
social_3012.set_index('datetime', drop=True, inplace=True)
social_3030.set_index('datetime', drop=True, inplace=True)

social_all = pd.concat([social_3005, social_3010, social_3011, social_3012, social_3030], axis=1).sort_index(axis = 0)
social_all = social_all.fillna(0)
social_mean = social_all.mean(axis=1, level=0)
social_std = social_all.std(axis=1, level=0)
social_mean = social_mean.rename(columns={'calllog': 'others_calllog', 'messagelog': 'others_messagelog', 'phonelog': 'others_phonelog', 'snslog': 'others_snslog', 'snsprop': 'others_snsprop'})
social_std = social_std.rename(columns={'calllog': 'others_calllog_std', 'messagelog': 'others_messagelog_std', 'phonelog': 'others_phonelog_std', 'snslog': 'others_snslog_std', 'snsprop': 'others_snsprop_std'})

social_others = pd.concat([social_mean, social_std], axis=1)
social_others

social_3012 = social_3012.rename(columns={'calllog': 'p3012_calllog', 'messagelog': 'p3012_messagelog', 'phonelog': 'p3012_phonelog', 'snslog': 'p3012_snslog', 'snsprop': 'p3012_snsprop'})
social_agg = pd.concat([social_3012, social_others], axis=1)
social_agg
```

```python
social_agg['z_calllog'] = (social_agg['p3012_calllog'] - social_agg['others_calllog']) / social_agg['others_calllog_std']
social_agg['z_messagelog'] = (social_agg['p3012_messagelog'] - social_agg['others_messagelog']) / social_agg['others_messagelog_std']
social_agg['z_phonelog'] = (social_agg['p3012_phonelog'] - social_agg['others_phonelog']) / social_agg['others_phonelog_std']
social_agg['z_snslog'] = (social_agg['p3012_snslog'] - social_agg['others_snslog']) / social_agg['others_snslog_std']
social_agg['z_snsprop'] = (social_agg['p3012_snsprop'] - social_agg['others_snsprop']) / social_agg['others_snsprop_std']

social_agg['z_social'] = (social_agg['z_calllog'] + social_agg['z_messagelog'] + social_agg['z_snsprop']) / 3
social_agg = social_agg.fillna(0)
social_agg
```
