```python
# Filtering out based on threshold
# Filter data if - Disturbance level >= 1, Emotion change <= -2, Emotion change >= 2

survey_UID = survey_UID.loc[(survey_UID.Disturbance_level < 1)]
survey_UID = survey_UID.loc[(survey_UID.Emotion_change > -2) & (survey_UID.Emotion_change < 2)]

survey_UID
```

```python
# Taking average of each day

survey_UID_index = survey_UID.copy()

survey_UID_index['datetime'] = pd.to_datetime(survey_UID_index['responseTime_KDT'])
survey_UID_index.set_index('datetime', drop=True, inplace=True) 
survey_UID_index_resampled = survey_UID_index.resample('1D').mean()
survey_UID_index_resampled = survey_UID_index_resampled[['UID', 'Valence', 'Arousal', 'Attention', 'Stress']]
survey_UID_index_resampled['Stress'] = (-survey_UID_index_resampled['Stress'])
survey_UID_index_resampled = survey_UID_index_resampled.drop('UID', axis=1)
survey_UID_index_resampled['total_mean'] = survey_UID_index_resampled.mean(axis=1)

survey_UID_index_resampled
```

## For z score extraction

```python
import pandas as pd 

mental_3012 = pd.read_csv('/content/drive/My Drive/CS492 DP/Data processing/[final] mental/mental-p3012.csv').iloc[1:, :]
mental_others = pd.read_csv('/content/drive/My Drive/CS492 DP/Data processing/[final] mental/mental-others.csv')

# change column name
mental_3012 = mental_3012.rename(columns={'Valence': 'p3012_Valence', 'Arousal': 'p3012_Arousal', 'Attention': 'p3012_Attention', 'Stress': 'p3012_Stress', 'total_mean': 'p3012_mental_mean'})
mental_others = mental_others.rename(columns={'Unnamed: 0': 'datetime', 'Valence_mean': 'others_Valence', 'Arousal_mean': 'others_Arousal', 'Attention_mean': 'others_Attention', 'Stress_mean': 'others_Stress', 'total_mean': 'others_mental_mean', 'Valence_std': 'others_Valence_std', 'Arousal_std': 'others_Arousal_std', 'Attention_std': 'others_Attention_std', 'Stress_std': 'others_Stress_std', 'total_std': 'others_mental_std'})

# set index with datetime
mental_3012.set_index('datetime', drop=True, inplace=True)
mental_others.set_index('datetime', drop=True, inplace=True)

mental_agg = pd.concat([mental_3012, mental_others], axis=1)
mental_agg
```

```python
mental_agg['z_Valence'] = (mental_agg['p3012_Valence'] - mental_agg['others_Valence']) / mental_agg['others_Valence_std']
mental_agg['z_Arousal'] = (mental_agg['p3012_Arousal'] - mental_agg['others_Arousal']) / mental_agg['others_Arousal_std']
mental_agg['z_Attention'] = (mental_agg['p3012_Attention'] - mental_agg['others_Attention']) / mental_agg['others_Attention_std']
mental_agg['z_Stress'] = (mental_agg['p3012_Stress'] - mental_agg['others_Stress']) / mental_agg['others_Stress_std']

# mental_agg['z_mental'] = (mental_agg['p3012_mental_mean'] - mental_agg['others_mental_mean']) / mental_agg['others_mental_std']

mental_agg['z_mental'] = (mental_agg['z_Valence'] + mental_agg['z_Arousal'] + mental_agg['z_Attention'] + mental_agg['z_Stress']) / 4

mental_agg
```

```python
# for weekly extraction
import pandas as pd 

survey_3005 = pd.read_csv('/content/drive/My Drive/CS492 DP/Data processing/[final] mental/mental-p3005.csv')
survey_3010 = pd.read_csv('/content/drive/My Drive/CS492 DP/Data processing/[final] mental/mental-p3010.csv')
survey_3011 = pd.read_csv('/content/drive/My Drive/CS492 DP/Data processing/[final] mental/mental-p3011.csv')
survey_3012 = pd.read_csv('/content/drive/My Drive/CS492 DP/Data processing/[final] mental/mental-p3012.csv').iloc[1:, :]
survey_3030 = pd.read_csv('/content/drive/My Drive/CS492 DP/Data processing/[final] mental/mental-p3030.csv')

survey_3005.set_index('datetime', drop=True, inplace=True)
survey_3010.set_index('datetime', drop=True, inplace=True)
survey_3011.set_index('datetime', drop=True, inplace=True)
survey_3012.set_index('datetime', drop=True, inplace=True)
survey_3030.set_index('datetime', drop=True, inplace=True)

mental_all = pd.concat([survey_3005, survey_3010, survey_3011, survey_3012, survey_3030], axis=1)
mental_all = mental_all.fillna(0)
mental_all = pd.concat([mental_all.iloc[:, :5],mental_all.iloc[:, 5:10], mental_all.iloc[:, 10:15], mental_all.iloc[:, 15:20], mental_all.iloc[:, 20:]], axis=0)

others_Valence = mental_all['Valence'].mean(axis=0)
others_Valence_std = mental_all['Valence'].std(axis=0)

others_Arousal = mental_all['Arousal'].mean(axis=0)
others_Arousal_std = mental_all['Arousal'].std(axis=0)

others_Attention = mental_all['Attention'].mean(axis=0)
others_Attention_std = mental_all['Attention'].std(axis=0)

others_Stress = mental_all['Stress'].mean(axis=0)
others_Stress_std = mental_all['Stress'].std(axis=0)

data = [[others_Valence, others_Arousal, others_Attention, others_Stress, others_Valence_std, others_Arousal_std, others_Attention_std, others_Stress_std]]
mental_agg_week = pd.DataFrame(data, columns = ['others_Valence', 'others_Arousal', 'others_Attention', 'others_Stress', 'others_Valence_std', 'others_Arousal_std', 'others_Attention_std', 'others_Stress_std'])

mental_agg_week['p3012_Valence'] = mental_agg['p3012_Valence'].mean(axis=0)
mental_agg_week['p3012_Arousal'] = mental_agg['p3012_Arousal'].mean(axis=0)
mental_agg_week['p3012_Attention'] = mental_agg['p3012_Attention'].mean(axis=0)
mental_agg_week['p3012_Stress'] = mental_agg['p3012_Stress'].mean(axis=0)
mental_agg_week['p3012_mental_mean'] = mental_agg['p3012_mental_mean'].mean(axis=0)
mental_agg_week['others_mental_mean'] = mental_agg['others_mental_mean'].mean(axis=0)
mental_agg_week['others_mental_std'] = mental_agg['others_mental_mean'].std(axis=0)

mental_agg_week['z_Valence'] = (mental_agg_week['p3012_Valence'] - mental_agg_week['others_Valence']) / mental_agg_week['others_Valence_std']
mental_agg_week['z_Arousal'] = (mental_agg_week['p3012_Arousal'] - mental_agg_week['others_Arousal']) / mental_agg_week['others_Arousal_std']
mental_agg_week['z_Attention'] = (mental_agg_week['p3012_Attention'] - mental_agg_week['others_Attention']) / mental_agg_week['others_Attention_std']
mental_agg_week['z_Stress'] = (mental_agg_week['p3012_Stress'] - mental_agg_week['others_Stress']) / mental_agg_week['others_Stress_std']

mental_agg_week['z_mental'] = (mental_agg_week['z_Valence'] + mental_agg_week['z_Arousal'] + mental_agg_week['z_Attention'] + mental_agg_week['z_Stress']) / 4
# mental_agg_week['z_mental'] = (mental_agg_week['p3012_mental_mean'] - mental_agg_week['others_mental_mean']) / mental_agg_week['others_mental_std']
mental_agg_week.index.name = 'Total'

mental_agg_week
```
